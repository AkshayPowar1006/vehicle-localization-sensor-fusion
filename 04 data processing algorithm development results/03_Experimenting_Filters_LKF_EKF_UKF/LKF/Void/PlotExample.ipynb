```python
import numpy as np
import matplotlib.pyplot as plt

# Define data and error values
x = np.array(['Group A', 'Group B', 'Group C', 'Group D'])
y1 = np.array([3, 5, 2, 6])
y2 = np.array([4, 7, 1, 8])
y3 = np.array([2, 6, 3, 5])
y_err = np.array([0.5, 1.2, 0.8, 0.7])

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Add bar plots to the first subplot
ax1.bar(x, y1, yerr=y_err, capsize=5, color='tab:blue', label='Dataset 1')
ax1.bar(x, y2, yerr=y_err, capsize=5, color='tab:green', bottom=y1, label='Dataset 2')
ax1.set_title('Bar Plot Example')
ax1.set_xlabel('Groups')
ax1.set_ylabel('Values')
ax1.legend(loc='upper left')

# Add a horizontal line and error bars to the second subplot
ax2.axhline(y=y3.mean(), color='tab:red', linestyle='--', label='Mean')
ax2.bar(x, y3, yerr=y_err, capsize=5, color=['tab:orange', 'tab:blue', 'tab:green', 'tab:purple'], label='Dataset 3')
ax2.set_title('Bar Plot Example with Error Bars')
ax2.set_xlabel('Groups')
ax2.set_ylabel('Values')
ax2.legend()

# Adjust spacing between subplots and save the figure
fig.tight_layout()
plt.savefig('bar_plot_example.png')
plt.show()
```
