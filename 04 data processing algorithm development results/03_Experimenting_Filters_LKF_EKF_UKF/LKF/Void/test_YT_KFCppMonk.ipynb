```python
import numpy as np
from YT_KFCppMonk.py import KF 
import unittest
```

```python
class TestKF(unittest.TestCase):

    def test_can_construct_with_x_v(self):
        x = 20
        v = 0.5
        kf = KF(init_x = x, init_v = v)
        self.assertAlmostEqual(kf.x[0], x)
        self.assertAlmostEqual(kf.v[0], v)
```
