```python
# Prediction Equation
def predict(vec):
    A = np.array([[1, 0, 1, 0],[0, 1, 0, 1], [0, 0, 1, 0], [0, 0, 0, 1]])
    X = np.array([true_measurement[1:5,0]])
    
    B = np.array([[0.5, 0], [0, 0.5], [1, 0], [0, 1]])
    U = np.array([true_measurement[5:,0]])
    
    AX = np.dot(A,np.transpose(X))
    BU = np.dot(B,np.transpose(U))
    
    X_pred = AX + BU
    print(X_pred)
    
    return A
```

```python
# Predicted process covariance Matrix
def pred_cov_matrix(sx, sy, sxv, syv):
    P = np.array([[sx*sx, sx*sy, sx*sxv, sx*syv],
         [sy*sx, sy*sy, sy*sxv, sy*syv],
         [sxv*sx, sxv*sy, sxv*sxv, sxv*syv],
         [syv*sx, syv*sy, syv*sxv, syv*syv]])
    return P
```
