```python
import numpy as np
import math

def next(x,y,phi,phi_dot, omega):
    rad = 0.055
    time = 1
    x1 = x + time * omega * math.cos(phi) * rad 
    y1 = y + time * omega * math.sin(phi) * rad 
    phi = phi + phi_dot * time
    phi_dot = phi_dot
    
    print(np.array([x1,y1,phi,phi_dot, np.sqrt(np.square(x1-x)+np.square(y1-y))]))
```

```python
next(12.2953,51.2772,1.155,0.005,35.4)
```
