**Title: Extract .json to .csv**<br><br>
**Description:** This code reads the .json file generated from the Swift Navigation console. Afterward, it selects particular message types and retrieves relevant information such as latitude, longitude, and timestamps from them. This extracted data is then stored in a variable named "locationLogData" and exported as a .csv file that can be utilized for additional processing.<br><br>

```python
# Importing the necessary liabraries
import csv
import json
import numpy as np
import sys

# Define input and output file paths
in_file = "../Trial3/PiksiRTK/swift-gnss-20230414-134505.sbp.json"
out_file = "../Trial3/PiksiRTK/path5_1404023.csv"

# Load JSON data from input file into a list
data = []
with open(in_file) as jFile:
    for line in jFile:
        data.append(json.loads(line))

# Empty lists to store the extracted data
lat, lon, year, month, date, hours, minutes, seconds, ns= [], [], [], [], [], [], [], [], []
        
# Extract necessary data from JSON data        
for i in range(len(data)):
    if data[i]["msg_type"] == 522: # 522 has latitude and longitude data stored
        lat.append(data[i]["lat"])
        lon.append(data[i]["lon"])
        
    if data[i]["msg_type"] == 259: # 259 has timestamps data stored
        year.append(data[i]["year"])
        month.append(data[i]["month"])
        date.append(data[i]["day"])
        hours.append(data[i]["hours"])
        minutes.append(data[i]["minutes"])
        seconds.append(data[i]["seconds"])
        ns.append(data[i]["ns"])
        
# Combine extracted data into a 2D numpy array
locationLogData = np.transpose(np.array([year, month, date, hours, minutes, seconds, ns, lat, lon]))

# Write extracted data to output file in CSV format
with open(out_file, mode = 'w',newline='') as f:
    mywriter = csv.writer(f)
    for row in locationLogData:
        mywriter.writerow(row)
```
